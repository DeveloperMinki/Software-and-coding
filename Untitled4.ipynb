```python
import numpy as np
import cv2
cap = cv2.VideoCapture(0)
# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        frame = cv2.flip(frame,0)
        # write the flipped frame
        out.write(frame)
        cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    else:
        break
# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows() # 비디오 저장

```

```python
import cv2
import numpy as np
def nothing(x):
    pass
# Create a black image, a window
img = np.zeros((300,512,3), np.uint8)#이미지 빈 영역으로 지정
cv2.namedWindow('image')# 창 하나 만들기
# create trackbars for color change
cv2.createTrackbar('R','image',0,255,nothing)#트랙바크기
cv2.createTrackbar('G','image',0,255,nothing)
cv2.createTrackbar('B','image',0,255,nothing)
# create switch for ON/OFF functionality
switch = '0 : OFF \n1 : ON'
cv2.createTrackbar(switch, 'image',0,1,nothing)
while(1):
    cv2.imshow('image',img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
    # get current positions of four trackbars
    r = cv2.getTrackbarPos('R','image')
    g = cv2.getTrackbarPos('G','image')
    b = cv2.getTrackbarPos('B','image')
    s = cv2.getTrackbarPos(switch,'image')
    if s == 0:#오프면 아무것도 표시안함 이미지에
        img[:] = 0
    else:
        img[:] = [b,g,r]#해당값표시
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
BLUE = [255,0,0]
img1 = cv2.imread('opencv_logo.png')
replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)
plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
#끝컬러 연장
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
#반사 시킴
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')

plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
#바둑판식 배열
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```

```python
# Load two images
img1 = cv2.imread('test2.jpg')
img2 = cv2.imread('opencv_logo.png')
# I want to put logo on top-left corner, So I create a ROI
rows,cols,channels = img2.shape
roi = img1[0:rows, 0:cols ]
# Now create a mask of logo and create its inverse mask also
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)  #그래이로 바끔
ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)#글자부분
mask_inv = cv2.bitwise_not(mask)#색깔부분만 들고옴  색반전
# Now black-out the area of logo in ROI
img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv) #글자 투명 뒷 그림 보임
# Take only region of logo from logo image.
img2_fg = cv2.bitwise_and(img2,img2,mask = mask)#검정화면 로고가 흰색으로 바뀜
# Put logo in ROI and modify the main image
dst = cv2.add(img1_bg,img2_fg)#두개 합침
img1[0:rows, 0:cols ] = dst#최종 작업
cv2.imshow('res',img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
cap = cv2.VideoCapture(0)
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        cv2.imshow('frame',frame)
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        # define range of blue color in HSV
        lower_blue = np.array([21,50,50]) # hsv 컬러값
        upper_blue = np.array([41,255,255]) # hsv 컬러값
        # Threshold the HSV image to get only blue colors
        mask = cv2.inRange(hsv, lower_blue, upper_blue)
        # Bitwise-AND mask and original image
        res = cv2.bitwise_and(frame,frame, mask= mask)
        cv2.imshow('frame',frame)
        cv2.imshow('mask',mask)
        cv2.imshow('res',res)
    if cv2.waitKey(40) & 0xFF == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('test2.jpg',0)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
plt.figure(figsize=(17,10)) ## 크기 조절
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
h,w=256,512
img=np.zeros((h,w),np.uint8)
for i in range(h):
    for j in range(w):
        img[i,j]=(j)%256
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
#중앙값 127 127보다크면 흰색 작으면 검정
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
#중앙값 127 127보다크면 흰색 작으면 검정 반대
ret,thresh3 = cv2.threshold(img,60,255,cv2.THRESH_TRUNC)
#60까지 가다가 나머지는 흰색
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
#127까지 가다가 나머지는 검정 
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
#127까지 가다가 나머지는 검정 반대 
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
plt.figure(figsize=(17,10)) ## 크기 조절
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img = cv2.imread('lena.png', 0)
h,w=img.shape
dst = np.zeros((h-2*l,w-2*l),np.uint8)
for i in range(l,h-l):
    for j in range(l,w-l):
        t=0
        for p in range(-l,l+1):
            for q in range(-l,l+1):
                t += kernel[p+l,q+l]*img[i+p,j+q]
        dst[i-l,j-l]=t 
cv2.imshow('img',img)
cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
